```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('./dataset/yds_train2018.csv')
holidays = pd.read_excel('./dataset/holidays.xlsx')
promo = pd.read_csv('./dataset/promotional_expense.csv')
```

```python
# data.describe
data.keys
```

```python
data = data.drop(['S_No','Week','Merchant_ID'],axis=1)
data
```

```python
#shrink by performing groupby
data = pd.DataFrame(data.groupby(by=['Country','Year','Month','Product_ID'],sort=False)[['Sales']].sum()).reset_index()
data
```

```python
data['holiday'] = pd.Series(np.zeros(388))
data
```

```python
date_to_month = lambda x : int(x.split(',')[1].strip())
date_to_year = lambda x : int(x.split(',')[0].strip())
x = holidays['Date'].apply(date_to_month)
y = holidays['Date'].apply(date_to_year)
holidays['month'] = x
holidays['year'] = y
holidays
```

```python
holidays = holidays.drop(labels=['Date','Holiday'],axis=1)
holidays['count'] = pd.DataFrame(np.ones(652))
holidays
```

```python
holidays = pd.DataFrame(holidays.groupby(by=['Country','year','month'],sort=False)[['count']].sum()).reset_index()
holidays
```

```python
# holidays = holidays.values
holidays
```

```python
def get_holiday_count(country,year,month):
    for i in holidays:
        if(i[0]==country and i[1]==year and i[2]==month):
            return i[3]
    return 0
```

```python
holiday = np.zeros(388)
i=0
for row in data.iterrows():
    holiday[i] = get_holiday_count(row[1]['Country'],row[1]['Year'],row[1]['Month'])
    i+=1
data['holiday'] = pd.Series(holiday)
data.head()
```

```python
promo = promo.values
promo
```

```python
def get_expense(country,year,month,prod_id):
    for row in promo:
        if(row[2]==country and row[0]==year and row[1]==month and row[3]==prod_id):
            return row[4]
    return 0
```

```python
expense = np.zeros(388)
i = 0
for row in data.iterrows():
    expense[i] = get_expense(row[1]['Country'],row[1]['Year'],row[1]['Month'],row[1]['Product_ID'])
    i+=1
data['expense'] = pd.Series(expense)
data.head()
```
